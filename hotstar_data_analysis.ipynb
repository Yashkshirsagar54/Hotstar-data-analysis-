```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Clean Up Column Value

```python
print("Columns before:", list(df.columns))
```

```python
df.columns = df.columns.str.strip()
print("Columns after:", list(df.columns))
```

### handling missing data

```python
df = df.dropna(subset=['genre'])              # Drop rows missing genre
df = df.drop_duplicates()
```

```python
df.drop_duplicates()
```

```python
df.duplicated()
```

```python
df.describe()
```

### Split Multi-Genre Entries for Analysis

```python
df_expanded = (
    df['genre']
    .str.split(',', expand=True)
    .stack()
    .str.strip()
    .reset_index(level=1, drop=True)
    .to_frame('Genre')
)

df_expanded = df_expanded.join(df[['year', 'type']], how='left')
```

```python
genre_freq = df_expanded['Genre'].value_counts()
print(genre_freq)
```

### genre distribution

```python
genre_year = (
    df_expanded.groupby(['year', 'Genre'])
               .size()
               .unstack(fill_value=0)
)
print(genre_year)
```

### visualyzing the insights 

```python
sns.catplot(x="genre", data=df, hue="genre", legend=False, palette="Set2")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
sns.barplot(x=genre_freq.index, y=genre_freq.values, palette='viridis')
plt.xticks(rotation=45)
plt.title('Genre Frequency in Hotstar Dataset')
plt.ylabel('Number of Titles')
plt.show()

plt.figure(figsize=(14,8))
genre_year.plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Genre Trends Over Time')
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
